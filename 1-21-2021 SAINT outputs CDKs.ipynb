```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
combined_SAINT = pd.read_csv('/Users/plutzer/Box/CellBio-MajorLab/Users/Rotation Students/Isaac/SAINT outputs/CDKs_combined_SAINT.csv')
separate_SAINT = pd.read_csv('/Users/plutzer/Box/CellBio-MajorLab/Users/Rotation Students/Isaac/SAINT outputs/CDKs-separate_SAINT.csv')
```

```python
com_map = pd.read_csv('/Users/plutzer/Box/CellBio-MajorLab/Users/Rotation Students/Isaac/SAINT outputs/CDKs_combined_mapping.txt',delimiter='\t')
```

```python
sep_sub = separate_SAINT[['Bait','Prey','SaintScore']]
com_sub = combined_SAINT[['Bait','Prey','SaintScore']]
```

```python
merge = sep_sub.merge(com_sub,how='inner',on=['Bait','Prey'])
```

```python
len(merge)
```

```python
merge = merge.dropna()
```

```python
len(merge)
```

```python
fig, ax = plt.subplots(figsize = (15,10))
plt.scatter(merge['SaintScore_x'],merge['SaintScore_y'],s=16,c='green',alpha = 0.1)
ax.set_title('SAINT scores (separate vs. combined)',size = 18)
ax.set_ylabel('Combined',size = 14)
ax.set_xlabel('Separate',size = 14)
```

```python
merge['Difference'] = np.abs(merge['SaintScore_x']-merge['SaintScore_y'])
```

```python
merge[merge['Difference'] > 0.9]
```

```python
merge = merge[['CON' not in item for item in list(merge['Prey'])]]
merge = merge[['+' not in item for item in list(merge['Prey'])]]
merge = merge[['Q6ZSR9' not in item for item in list(merge['Prey'])]]
merge = merge[['REV' not in item for item in list(merge['Prey'])]]
merge = merge[['Q5SRD0' not in item for item in list(merge['Prey'])]]
```

```python


com_unis = list(com_map['From'])
com_genes = list(com_map['To'])
genes = []
preys = list(merge['Prey'])
for prey in preys:
    preylist = prey.split(';')
    genes.append(';'.join([com_genes[com_unis.index(key)] for key in preylist]))
    
```

```python
merge['Gene'] = genes
```

```python
cdk = ['CDK' in item for item in list(merge['Gene'])]
```

```python
cdk_merge = merge[cdk]
```

```python
fig, ax = plt.subplots(figsize = (15,10))
plt.scatter(cdk_merge['SaintScore_x'],cdk_merge['SaintScore_y'],s=16,c='green')
ax.set_title('CDK SAINT scores (separate vs. combined)',size = 18)
ax.set_ylabel('Combined',size = 14)
ax.set_xlabel('Separate',size = 14)
plt.savefig('CDKs_scatter.png',dpi=256)
```
