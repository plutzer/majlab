```python
dfs_list = []
for item in pathslist:
    dfs_list.append(pd.read_csv(item,delimiter='\t'))
```

```python
pg_merged = []
for item in dfs_list:
    pg_merged.append(list(item['Protein IDs']))
flat_pg_merged = []
for sublist in pg_merged:
    for item in sublist:
        flat_pg_merged.append(item)
```

```python
pg_1 = [] # pg_1 is the protein groups with duplicates dropped.
for item in flat_pg_merged:
    if item not in pg_1:
        pg_1.append(item)
```

```python
pg_2 = [] # pg_2 is the protein groups that don't share any elements with any other group
leftovers2 = [] #This is a protein group of non-unique groups
for index in range(len(pg_1)):
    proteins = pg_1[index].split(';')
    unique = True
    for protein in proteins:
        if protein in ';'.join(pg_1[:index] + pg_1[index+1:]):
            #print(pg_1[index])
            unique = False
    if unique:
        pg_2.append(pg_1[index])
    else:
        leftovers2.append(pg_1[index])
```

```python
contained_list = [] # Generate a list that denotes how many other groups are a subset of this protein group
for item1 in leftovers2:
    count = 0
    for item2 in leftovers2:
        if item2 in item1:
            count = count + 1
    contained_list.append(count)
```

```python
pg_3 = [item for item in pg_2] #pg_3 has all 
contained_list_3
vals_1 = [i for i, x in enumerate(contained_list) if x == 1]
vals_2 = [i for i, x in enumerate(contained_list) if x != 1]
leftovers3 = []
for index in vals_1:
    pg_3.append(leftovers2[index])
for index in vals_2:
    leftovers3.append(leftovers2[index])
    contained_list_3.append(contained_list[index])
```
