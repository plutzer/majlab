```python
import pandas as pd
import numpy as np
```

```python
saint_df = pd.read_csv('/Users/plutzer/Box/CellBio-MajorLab/Users/Rotation Students/Isaac/IDG outputs and ED files/DhavalBaits-Merge_CompPASS_SAINT.csv')
mapping_df = pd.read_csv('/Users/plutzer/majlab/megan_mapping.txt',delimiter='\t')
```

```python
uniprots = list(saint_df['PreyGene'])
```

```python
mapping_dict = mapping_df.set_index('From').to_dict()['To']
```

```python
genes = []
for item in uniprots:
    try:
        ids = item.split(';')
        group = []
        for uniprot_id in ids:
            try:
                group.append(mapping_dict[uniprot_id])
            except:
                group.append('No_mapping')
        gene_group = ';'.join(group)
    except:
        gene_group = 'NaN'
    genes.append(gene_group)
```
